```python
import pandas as pd
import matplotlib.pyplot as plt

# Tạo DataFrame dữ liệu cây xanh tại châu Phi
data = {
    'Năm': [2015, 2015, 2015, 2016, 2016, 2016, 2017, 2017, 2017,
            2018, 2018, 2018, 2019, 2019, 2019, 2020, 2020, 2020],
    'Quốc gia': ['Nigeria', 'Kenya', 'Ethiopia'] * 6,
    'Số cây được trồng (triệu cây)': [
        12.5, 8.3, 15.1,
        14.2, 9.5, 18.0,
        17.0, 11.3, 22.5,
        20.4, 13.2, 25.8,
        23.1, 15.0, 28.4,
        26.8, 17.2, 30.9
    ]
}

df = pd.DataFrame(data)

# Hiển thị dữ liệu
print("Dữ liệu cây xanh tại châu Phi:")
print(df)
```

```python
# -----------------------------
# Biểu đồ 1: Biểu đồ đường (theo thời gian)
# -----------------------------
plt.figure(figsize=(8, 5))
for country in df['Quốc gia'].unique():
    subset = df[df['Quốc gia'] == country]
    plt.plot(subset['Năm'], subset['Số cây được trồng (triệu cây)'], marker='o', label=country)

plt.title("Tổng số cây xanh được gieo trồng tại Châu Phi (2015–2020)")
plt.xlabel("Năm")
plt.ylabel("Số cây được trồng (triệu cây)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -----------------------------
# Biểu đồ 2: Biểu đồ cột nhóm
# -----------------------------
pivot_df = df.pivot(index='Năm', columns='Quốc gia', values='Số cây được trồng (triệu cây)')
pivot_df.plot(kind='bar', figsize=(9, 5))

plt.title("So sánh số cây được trồng giữa các quốc gia Châu Phi")
plt.xlabel("Năm")
plt.ylabel("Số cây được trồng (triệu cây)")
plt.legend(title='Quốc gia')
plt.grid(axis='y')
plt.show()
```

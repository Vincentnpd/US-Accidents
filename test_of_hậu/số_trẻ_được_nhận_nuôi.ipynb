```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# TẠO DỮ LIỆU MẪU
# -----------------------------
data = {
    'Năm': [2015, 2015, 2015, 2016, 2016, 2016, 2017, 2017, 2017, 2018, 2018, 2018,
            2019, 2019, 2019, 2020, 2020, 2020, 2021, 2021, 2021, 2022, 2022, 2022,
            2023, 2023, 2023, 2024, 2024, 2024],
    'Quốc gia': ['Mỹ', 'Việt Nam', 'Pháp'] * 10,
    'Số trẻ em được nhận nuôi': [
        125000, 2500, 7000,
        128000, 2600, 6800,
        130000, 2750, 7200,
        132500, 2800, 7300,
        136000, 2900, 7400,
        138000, 2950, 7600,
        140000, 3000, 7800,
        145000, 3100, 8000,
        150000, 3200, 8200,
        155000, 3300, 8300
    ]
}

df = pd.DataFrame(data)

# Hiển thị dữ liệu
print("=== DỮ LIỆU NHẬN NUÔI TRẺ EM (2015–2024) ===")
print(df)

# -----------------------------
# BIỂU ĐỒ 1: ĐƯỜNG THEO THỜI GIAN
# -----------------------------
plt.figure(figsize=(8, 5))
for country in df['Quốc gia'].unique():
    subset = df[df['Quốc gia'] == country]
    plt.plot(subset['Năm'], subset['Số trẻ em được nhận nuôi'], marker='o', label=country)

plt.title("Tổng số trẻ em được nhận nuôi (2015–2024)")
plt.xlabel("Năm")
plt.ylabel("Số trẻ em được nhận nuôi")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------------
# BIỂU ĐỒ 2: CỘT NHÓM THEO NĂM
# -----------------------------
pivot_df = df.pivot(index='Năm', columns='Quốc gia', values='Số trẻ em được nhận nuôi')
pivot_df.plot(kind='bar', figsize=(9, 5))

plt.title("So sánh số trẻ em được nhận nuôi theo quốc gia (2015–2024)")
plt.xlabel("Năm")
plt.ylabel("Số trẻ em được nhận nuôi")
plt.legend(title="Quốc gia")
plt.grid(axis='y')
plt.show()

# -----------------------------
# BIỂU ĐỒ 3: TỔNG HỢP THEO NĂM
# -----------------------------
yearly_total = df.groupby('Năm')['Số trẻ em được nhận nuôi'].sum().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(yearly_total['Năm'], yearly_total['Số trẻ em được nhận nuôi'])
plt.title("Tổng số trẻ em được nhận nuôi trên toàn cầu (2015–2024)")
plt.xlabel("Năm")
plt.ylabel("Tổng số trẻ em được nhận nuôi")
plt.grid(axis='y')
plt.show()
```
